```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 11)
y = x ** 2

plt.plot(x, y, 'red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.2, 0.2, 0.7, 0.7]) #esquerda, inferior, largura, altura
axes2 = fig.add_axes([0.3, 0.4, 0.5, 0.3])

axes.plot(x, y, 'blue')
axes.set_xlabel("Eixo X")
axes.set_ylabel("Eixo Y")
axes.set_title("Título")

axes2.plot(y, x, 'red')
axes2.set_xlabel("Eixo X2")
axes2.set_ylabel("Eixo Y2")
axes2.set_title("Título2")
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'r--')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for axis in ax:
    axis.plot(x, y, 'r')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

ax[0].plot(x, y, 'red') #funciona como matriz se for 2 row é [][]
ax[1].plot(x, y, 'green') #funciona como matriz se for 2 row é [][]
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, color='#FF50AB', alpha=1, label='X**2')
ax.plot(x, x**3, 'g*-', alpha=0.5, label='X**3')

ax.set_title("Título")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")

ax.legend(loc='center left')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo Eixos padrão')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight')
axes[1].set_title('Intervalo Eixos apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 4])
axes[2].set_title('Intervalo Eixos cortados')
```
